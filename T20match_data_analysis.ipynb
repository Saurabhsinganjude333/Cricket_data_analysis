<h1 align="center"> T20 World Cup Cricket Data Cleaning </h1>

```python
#import necessary libraries
import pandas as pd
import json
```

<h4 style="color:blue">(1) Process Match Results</h4>

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

**Use scorecard as a match id to link with other tables**

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

**Create a match ids dictionary that maps team names to a unique match id. **

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

<h4 style="color:blue">(2) Process Batting Summary</h4>

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

**Cleanup weird characters**

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)
```

<h4 style="color:blue">(3) Process Bowling Summary</h4>

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

<h4 style="color:blue">(4) Process Players Information</h4>

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

**Cleanup weird characters**

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv('t20_csv_files/dim_players_no_images.csv', index = False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df_match = pd.read_csv('t20_csv_files/dim_match_summary.csv')
df_batting = pd.read_csv('t20_csv_files/fact_bating_summary.csv')
df_bowling = pd.read_csv('t20_csv_files/fact_bowling_summary.csv')
df_players = pd.read_csv('t20_csv_files/dim_players_no_images.csv')
```

# Perform analytics and visualization

### Example 1: Distribution of runs scored by batting position

```python
sns.boxplot(x='battingPos', y='runs', data=df_batting)
plt.title('Distribution of Runs Scored by Batting Position')
plt.xlabel('Batting Position')
plt.ylabel('Runs')
plt.show()
```

### Example 2: Top 10 bowlers with the highest number of wickets

```python
top_bowlers = df_bowling.groupby('bowlerName')['wickets'].sum().nlargest(10)
top_bowlers.plot(kind='barh', color='skyblue')
plt.title('Top 10 Bowlers with Highest Number of Wickets')
plt.xlabel('Number of Wickets')
plt.ylabel('Bowler')
plt.gca().invert_yaxis()
plt.show()
```

### Example 3: Comparison of batting styles

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='battingStyle', data=df_players)
plt.title('Distribution of Batting Styles')
plt.xlabel('Batting Style')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Example 4: Distribution of playing roles

```python
plt.figure(figsize=(8, 6))
sns.countplot(y='playingRole', data=df_players, order=df_players['playingRole'].value_counts().index)
plt.title('Distribution of Playing Roles')
plt.xlabel('Count')
plt.ylabel('Playing Role')
plt.show()
```

### Performance of Top Batsmen

```python
top_batsmen = df_batting.groupby('batsmanName')['runs'].sum().nlargest(10)
top_batsmen.plot(kind='bar', color='skyblue')
plt.title('Top Batsmen in T20 World Cup')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

### Player Roles Distribution by Team

```python
# Define the top teams
top_teams = ['India', 'Australia', 'England', 'Pakistan', 'West Indies']

# DataFrame for top teams
df_players_top_teams = df_players[df_players['team'].isin(top_teams)]

plt.figure(figsize=(12, 8))
sns.countplot(y='playingRole', hue='team', data=df_players_top_teams, 
              order=df_players_top_teams['playingRole'].value_counts().index)
plt.title('Player Roles Distribution by Team (Top Teams Only)')
plt.xlabel('Count')
plt.ylabel('Playing Role')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Analyzing player contributions/ Top Players Performance Metrics

### Top 5 highest Runs Scorer

```python
# Calculate total runs scored by each player
player_runs = df_batting.groupby('batsmanName')['runs'].sum().sort_values(ascending=False)

# Filter top 5 players based on total runs scored
top_runs_players = player_runs.head(5)

# Plot the total runs scored by the top 5 players
top_runs_players.plot(kind='bar', figsize=(12, 8), color='skyblue')
plt.title('Top 5 Players by Total Runs Scored')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.show()


```

### Top 5 Players by Total Wickets Taken

```python
# Calculate total wickets taken by each player
player_wickets = df_bowling.groupby('bowlerName')['wickets'].sum().sort_values(ascending=False)

# Filter top 5 players based on total wickets taken
top_wickets_players = player_wickets.head(5)

# Plot the total wickets taken by the top 5 players
top_wickets_players.plot(kind='bar', figsize=(12, 8), color='skyblue')
plt.title('Top 5 Players by Total Wickets Taken')
plt.xlabel('Player')
plt.ylabel('Total Wickets')
plt.show()
```

### Effect of Winning Toss on Match Outcome

```python
toss_outcome = df_match[df_match['team1'] == df_match['winner']].groupby('team1')['team1'].count()
toss_outcome.plot(kind='bar', figsize=(12, 8), color='skyblue')
plt.title('Effect of Winning Toss on Match Outcome')
plt.xlabel('Team')
plt.ylabel('Number of Wins after Winning Toss')
plt.show()
```

### Number of Matches Won by Each Team


```python
team_wins = df_match['winner'].value_counts()
plt.figure(figsize=(10, 6))
team_wins.plot(kind='bar', color='skyblue')
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```
